### Importing and cleaning 538's twitter troll data found here:
https://github.com/fivethirtyeight/russian-troll-tweets

#### Currently this script will combine the 9 files into a single dataset, and then subset it to english language and left or right trolls with 20 or more tweets

```python
import pandas as pd
```

```python
list_ = []
for i in range(1,10):
    df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/russian-troll-tweets/master/IRAhandle_tweets_'
                       + str(i) + '.csv')
    list_.append(df)

trolls = pd.concat(list_)

trolls.head()
```

```python
trolls.account_type.unique()
```

### Subset to english language tweets, no retweets, left and right trolls only

```python
trolls = trolls[(trolls['language'] == 'English') & (trolls['retweet'] == 0) & 
             (trolls['account_category'].isin(['RightTroll','LeftTroll']))]
trolls = trolls.drop(columns = ['language', 'retweet', 'account_type', 'post_type', ])
trolls.head()
```

```python
len(trolls.author.unique())
```

### Subset to trolls with over 200 posts

```python
# creating a mask for accounts with 200 or more tweets
big_trolls = pd.DataFrame(trolls.groupby('author').count()['content'])
big_trolls = big_trolls[big_trolls['content'] >= 20]
print(big_trolls.shape)
big_trolls.head()
```

```python
power_trolls = trolls[trolls['author'].isin(big_trolls.index)]
print(power_trolls.shape)
power_trolls.head()
```

### Export to .csv

```python
power_trolls.to_csv('troll_tweets.csv')
```
